```python
#6.
#（1）设置坐标轴的显示参数，x轴，y轴,是什么变量，刻度是多少，把坐标轴的数字刻度改为文字刻度并改变它的字体
#（2）修改坐标轴的位置，改变坐标轴的原点，也就是整体移动坐标轴
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1,1,10)
y = 2*x + 2                             #2乘x
y1 = x**2 + 2                           #平方用’**‘表示，在python中
y2 = x**3

plt.plot(x,y,c='C2',linestyle='--')
plt.plot(x,y1,c='C5',linestyle='-.')
plt.plot(x,y2,c='C8',linestyle=':')

plt.xlim(-2,2)                          #设置x轴的起始位置，也是limit限制
plt.ylim(-2,2)                          #设置y轴的起始位置，也是limit限制
plt.xlabel('sjjs')                      #设置x轴的名称
plt.ylabel("jjjjj")                     #设置y轴的名称

new_ticks = np.linspace(-5,5,10)        #在-5到5之间等分的选取10个点           
plt.xticks(new_ticks)                   #x轴的数字将会发生改变，变为-5,5上的10个点
plt.yticks(new_ticks)                   #y轴的数字将会发生改变，变为-5,5上的10个点
plt.xticks([-4,-2,0,2,4],['really bad','$bad\\alpha$','$normal$','$good$','$really\ good$'])
#坐标轴上文字的字，以上面为例体,加$--$,是将字体改为斜体
#命令：plt.xticks([-2,-1,0,1,2],['$really bad$','$bad$','$normal$','$good$','$really good$'])
#前提是设置字体的时候，如果要读出’空格‘，需要加入转字符“\”,才可以时空格读出，
#在设置字体时候要打出数学形式的符号例如’α‘，需要转字符’\‘，命令：$bad\\alpha$
# 不再设置字体的情况下，不需要转字符


ax =plt.gca()               #gca = 'get current axis'  把现在的这个轴拿出来,画出来的这张图的坐标轴，用ax来表示.
ax.spines['right'].set_color('none')    #spines代表坐标轴的骨架，right，left，top，bottom（上，下，左，右）
ax.spines['top'].set_color('none')
#到此，图上的x，y，轴是哪一个还没有定义，下面才来正式定义。
ax.xaxis.set_ticks_position('bottom')           #把x轴定义为‘bottom’的那条线的位置
ax.yaxis.set_ticks_position('left')             #把y轴定义为‘left'的那条线的位置
ax.spines['bottom'].set_position(('data',0))    #把’bottom‘（也就是x轴）移动到y=0的位置
ax.spines['left'].set_position(('data',0))      #把’left‘（也就是y轴）移动到x=0的位置


plt.show()
```

```python
#11
#柱状图，bar
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(n)    #X是随机生成的数字
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)  #Y1，Y2，是根据X的值生成的0.5到1.0的12个数字
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')    #绘制柱状图，bar，‘+’表示x轴的上方，‘-‘表示x轴的下方
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')    #facecolor，表示填充的颜色；edgecolor表示边框的颜色

for  x,y in zip(X,Y1):
    plt.text(x + 0.1,y + 0.05,'%.2f'%y,ha='center',va='bottom')
#ha:horizontal alignment,水平对齐；vertical alignment：垂直对齐
for x,y in zip(X,Y2):
    plt.text(x + 0.1,-y - 0.05,'-%.2f'%y,ha='center',va='top')
    


plt.show()
```

```python
#12.
#等高线图，contour
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    #the hight function，根据x和y值通过此函数来得到每一点的高度，高度是通过网格的升高来得到的，用到了meshgrid函数
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)


n = 256
x = np.linspace(-3,3,n) #用来得到x值
y = np.linspace(-3,3,n) #用来得到y值
X,Y = np.meshgrid(x,y)  #通过x,y值来计算高度，从而得到网格的高度，把x,y绑定成网格的输入值，相当于z

#使用 plt.contourf 目的是 填入等高线（contours）
plt.contourf(X,Y,f(X,Y),10,alpha=0.5,cmap=plt.cm.cool)    
#plt出等高线示意图，contourf,plt.contorf(x,y,z),cmap是填充颜色

C = plt.contour(X,Y,f(X,Y),10,colors='k',lw=.5)   #作用是对于等高线的参数进行设置

plt.clabel(C,inline=True,fontsize=10)                    #作用是在等高线上写上高度，也就是数字


plt.xticks(())    #作用是将x轴的坐标去掉
plt.yticks(())    #作用是将y轴的坐标去掉
plt.show()
```

```python
#14.
#3D图像需要加入一个模块，mpl.toolkits.mplot3d ，使用ax.plot_surface（）命令来画3D
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()  #创建一个figure,对于画二维图来说如果仅有一个图的话，此条可以不用写出，如果有多个图，需要写上。
                     #如果对以一个三维图来说，不写这条，图片画不出来
ax =Axes3D(fig)     #在图上加一个ax也就是轴，变成3D，3个轴

#x,y值
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)       #和等高线图中的一样
R = np.sqrt(X**2+Y**2)       #z的高度值，是通过X,Y来计算出来

#height value
Z = np.sin(R)                #z的值

ax.plot_surface(X,Y,Z,rstride=10,cstride=1,cmap=plt.get_cmap('rainbow'))
#ax,plot_surface是画3D图的方式，X,Y,Z三个坐标轴，一个是横跨度，一个是纵跨度，cmap是三维图的形状，可以在文档中查看
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')
#这条命令是用来画等高线的，zdir是从哪个方向看过去的视角，此图是从z轴看过去的，offset是呈现在哪个平面上，此图是在z=-2平面上
ax.set_zlim(-2,2)       #z轴的范围是-2到2

plt.show()
```

```python
#15.（3）
###python中都是从0开始的，并不是从1开始
#相比于上面的2种方法，下面这几种方法相对于好理解一点，使用subplot2grid。
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure()

ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)  #对于3行3列,(0,0)从第0行第0列开始，横向跨度是3,纵向跨度是1
ax1.plot([1,2],[2,3])
ax1.set_title('ax1_title')

ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)   #对于3行3列,(1,0)从第1行第0列开始，横向跨度是2,纵向跨度是1
ax2.scatter([5,3],[8,9])

ax3 = plt.subplot2grid((3,3),(1,2),colspan=1,rowspan=2)   #对于3行3列,(1,2)从第1行第2列开始，横向跨度是1,纵向跨度是2
ax3.plot([3,2],[5,8])

ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,rowspan=1)   #对于3行3列,(2,0)从第2行第0列开始，横向跨度是1,纵向跨度是1

ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,rowspan=1)   #对于3行3列,（2,1)从第2行第1列开始，横向跨度是1,纵向跨度是1

plt.tight_layout()
plt.show()
```

```python
#15.(4)
#使用gridspec，采用索引的方式
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure()
gs = gridspec.GridSpec(3,3)

ax1 = plt.subplot(gs[0,0:3])  #第0行，0到3列
ax2 = plt.subplot(gs[1,0:2])  #第1行，0到2列
ax3 = plt.subplot(gs[1:3,2])  #第1到2行，第2列
ax4 = plt.subplot(gs[-1:3,0]) #不懂
ax5 = plt.subplot(gs[-1:3,-2]) #不懂

plt.tight_layout()              #这条命令是时一个图上的个个字不受遮挡，看得更清楚
plt.show()
```

```python
#15.（6）
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
x = [1,2,3,4,5,6,7]
y = [1,5,3,7,13,18,15]


left,bottom,width,heigh = 0.1,0.1,0.8,0.8
ax1 = fig.add_axes([left,bottom,width,heigh])
ax1.plot(x,y,C='C3')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

left,bottom,width,heigh = 0.2,0.6,0.25,0.25
ax2 = fig.add_axes([left,bottom,width,heigh])
ax2.plot(y,x,'b')
ax2.set_xlabel('ww')
ax2.set_ylabel('y')
ax2.set_title('title2')

left,bottom,height,width = 0.6,0.2,0.25,0.25
ax3 = fig.add_axes([left,bottom,heigh,width])
ax3.plot(y,x,c='C5')
ax3.set_xlabel('rrr')
ax3.set_ylabel('y')
ax3.set_title('title3')


plt.show()
```

```python
#16一个图上有主次坐标轴，共享一个x轴，左右2侧分别是2个不同的y轴
import matplotlib.pyplot as plt 
import numpy as np


x = np.arange(0,10,0.1)
y1 = 0.05*x**2
y2 = -1*y1

fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x,y1,'g-')
ax2.plot(x,y2,'b--')

ax1.set_xlabel('x data')
ax1.set_ylabel('y1',color='g')
ax2.set_ylabel('y2',color='b')

plt.show()
```
